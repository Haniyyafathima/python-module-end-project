```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the excel file
df = pd.read_excel('C:/Users/haniy/Downloads/myexcel.xlsx')
df.head(10)
```

## Preprocessing

```python

# Replace the Height column with random integers between 150 and 180

df["Height"] = np.random.randint(150, 181, size=len(df))

```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Salary'].median()
```

```python
# Fill missing values in "Salary" with the median salary

df.fillna({'Salary':df['Salary'].median()}, inplace=True)
```

```python
# Fill missing values in "College" with 'Unknown'

df.fillna({'College':'Unknown'}, inplace=True)
```

## Analysis task

## Task 1:Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. 

```python
# Count the number of employees in each team
team_distribution = df['Team'].value_counts()
# Calculate the percentage split
team_percentage = (team_distribution / len(df)) * 100

team_distribution = pd.DataFrame({'Employee count':team_distribution,'Percentage':team_percentage})
team_distribution
```

```python
# visualization
plt.figure(figsize=(10,8))

sns.barplot(x=team_distribution.index,y=team_distribution["Employee count"], palette="coolwarm")
 
plt.title('Employee Distribution Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.show()
```

## Task 2 :Segregate employees based on their positions within the company

```python
position_dist = df['Position'].value_counts()

position_df = pd.DataFrame({'Position': position_dist.index, 'Count': position_dist.values})
print(position_df)
```

```python

# Visualisation

plt.figure(figsize=(10,5))
sns.barplot(x=position_dist.index, y=position_dist.values,palette='YlOrBr')
plt.xlabel('Position of Employees')
plt.ylabel('Number of Employees')
plt.title('Employee Distribution across Position')
plt.show()
```

## Task 3:Identify the predominant age group among employees

```python
# Define age bins and labels
bins = [20, 25, 30, 35, 40,45]
labels = ['20-25', '26-30', '31-35', '36-40', '41-45']

# Create a new column for age group
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_grp_counts = df['Age Group'].value_counts()
pd.DataFrame(age_group_counts)
```

```python
# Visualization for age across Age Group

plt.figure(figsize=(8,5))
sns.barplot(x = age_grp_counts.index, y= age_grp_counts.values,hue=age_grp_counts.values,palette='bright')
plt.xlabel('Age Group of Employees')
plt.ylabel('Count in the Age group')
plt.title('Age Distribution across Age Group')
```

 ## Task 4. Highest Salary Expenditure by Team & Position

```python
# 4. Discover which team and position have the highest salary expenditure.

# finding the Highest Salary Expenditure per Team

team_salary_expenditure = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)

#Display the top 5 teams
print(team_salary_expenditure.head())
print("The team with highest salary expenditure:", team_salary_expenditure.idxmax())
```

```python
# Visualization for the Highest Salary Expenditure per Team

plt.figure(figsize=(10,6))
sns.barplot(x=team_salary_expenditure.index,y=team_salary_expenditure.values, palette="coolwarm")
plt.xlabel("Team")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Team")
plt.xticks(rotation=90)
plt.show()
```

```python
# finding the Salary Expenditure per Position

position_sal_expen = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

#Display the top 5 position
print(position_sal_expen.head())
print("The position with highest salary expenditure:", position_sal_expen.idxmax())
```

```python
# Visualization for the Highest Salary Expenditure per Position

plt.figure(figsize=(10,8))
df.groupby("Position")["Salary"].sum().plot(kind="pie", autopct="%1.1f%%", cmap="coolwarm", startangle=90)
plt.title("Salary Expenditure Distribution by Position")
plt.ylabel("")  
plt.show()
```

## Task 5 : Investigate if there's any correlation between age and salary

```python
# Checking Correlation
correlation = df["Age"].corr(df["Salary"])
print('correlation between age and salary:',correlation)
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Age'], y=df['Salary'], color='red')
sns.regplot(x=df["Age"], y=df["Salary"], scatter=False, color="blue")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title('Correlation between Age and Salary')
plt.grid()
plt.show()
```
